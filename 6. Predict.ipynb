```python
import pandas as pd
import numpy as np
from usfull_tools import load_DS
#from catboost import Pool, CatBoostRegressor, CatBoostClassifier
#from sklearn.model_selection import train_test_split, cross_val_score
pd.options.display.max_columns = None
%matplotlib inline

from set_vars import KAGGLE_PREFIX, debug_mode, KAGGLE_DIR, target_column, target_type, id_column

train, test = load_DS(debug_mode, KAGGLE_DIR, KAGGLE_PREFIX, '_prepare.csv')
del train

cols = pd.read_csv(KAGGLE_DIR + KAGGLE_PREFIX + '_important_columns.csv')
test = test[cols['Feature']]
test.shape
```

```python
import pickle

with open(KAGGLE_DIR + KAGGLE_PREFIX + 'classifier.pkl', 'rb') as fid:
    model = pickle.load(fid)

y_test = model.predict(test)
predict = pd.read_csv(KAGGLE_DIR + KAGGLE_PREFIX + 'test.csv', usecols=[id_column])
predict[target_column] = y_test.astype(np.int)
predict.to_csv(KAGGLE_DIR + KAGGLE_PREFIX + 'submission.csv', index=False, sep=',')
```
